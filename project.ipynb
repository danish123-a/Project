```
from typing import List, Dict
import subprocess

class TaskManager:
    def __init__(self, scope: List[str], wildcard_domains: List[str]):
        self.scope = scope
        self.wildcard_domains = wildcard_domains
        self.task_list = []

    def add_task(self, task: Dict):
        """Add a task to the task list."""
        self.task_list.append(task)

    def execute_task(self, task: Dict):
        """Execute a task using subprocess."""
        tool = task.get("tool")
        command = task.get("command")

        try:
            result = subprocess.run(command, shell=True, capture_output=True, text=True)
            if result.returncode == 0:
                return result.stdout
            else:
                raise Exception(result.stderr)
        except Exception as e:
            print(f"Task failed: {e}")
            return None

    def run(self):
        """Execute all tasks in the task list."""
        for task in self.task_list:
            output = self.execute_task(task)
            if output:
                print(f"Task completed: {task['description']}")
                self.process_output(output)
            else:
                print(f"Task failed: {task['description']}")

    def process_output(self, output: str):
        """Process tool output and add new tasks if needed."""
        if "nmap" in output:
            # Example: Add a gobuster task for discovered subdomains
            self.add_task({
                "tool": "gobuster",
                "command": "gobuster dir -u example.com -w /path/to/wordlist.txt",
                "description": "Directory brute-forcing on example.com"
            })
```

```
class ScopeEnforcer:
    def __init__(self, scope: List[str], wildcard_domains: List[str]):
        self.scope = scope
        self.wildcard_domains = wildcard_domains

    def is_within_scope(self, target: str) -> bool:
        """Check if a target is within the defined scope."""
        if target in self.scope:
            return True
        for domain in self.wildcard_domains:
            if target.endswith(domain):
                return True
        return False
```

```
class ToolExecutor:
    @staticmethod
    def run_nmap(target: str):
        command = f"nmap -sV {target}"
        return {"tool": "nmap", "command": command, "description": f"Port scanning on {target}"}

    @staticmethod
    def run_gobuster(target: str, wordlist: str):
        command = f"gobuster dir -u {target} -w {wordlist}"
        return {"tool": "gobuster", "command": command, "description": f"Directory brute-forcing on {target}"}
```

```
class FailureHandler:
    def __init__(self, max_retries: int = 3):
        self.max_retries = max_retries

    def retry_task(self, task: Dict, retry_count: int):
        if retry_count < self.max_retries:
            print(f"Retrying task: {task['description']}")
            return True
        else:
            print(f"Task failed after {self.max_retries} retries: {task['description']}")
            return False
```

```
import logging
from datetime import datetime

class Logger:
    def __init__(self):
        logging.basicConfig(filename="cybersecurity_agent.log", level=logging.INFO)

    def log(self, message: str):
        logging.info(f"{datetime.now()} - {message}")

class Reporter:
    def generate_report(self, task_list: List[Dict]):
        report = "Cybersecurity Audit Report\\n"
        report += "========================\\n"
        for task in task_list:
            report += f"{task['description']} - Status: {task.get('status', 'Unknown')}\\n"
        return report
```
